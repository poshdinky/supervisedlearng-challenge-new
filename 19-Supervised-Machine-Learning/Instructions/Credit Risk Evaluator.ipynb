```python
#import libraries:

import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.datasets import make_regression
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
train_df = pd.read_csv('Resources/2019loans.csv')
test_df = pd.read_csv('Resources/2020Q1loans.csv')
train_df
```

```python
X_train = train_df.drop(['loan_status'], axis = 1)
X_train = pd.get_dummies(X_train)

y_train = train_df['loan_status']
```

```python
X_test = test_df.drop(['loan_status'], axis = 1)
X_test = pd.get_dummies(X_test)

y_test = test_df['loan_status']
```

```python
missing_cols = set( X_train.columns ) - set( X_test.columns )
# Add a missing column in test set with default value equal to 0
for c in missing_cols:
    X_test[c] = 0
# Ensure the order of column in the test set is in the same order than in train set
X_test = X_test[X_train.columns]
```

## Prediction : Consider the models

the machine learning algorithms Random Forest and Logistic Regression are used for various disciplines for classification and regression purposes, Logistic regression is a parameter based model and random forest is a non parametric model.
There is no single algorithm which performs best for all possible datasets,and there is no model better than the other.
For example let's take the algorithms that we will work with in this assignment, logistic regression works better when we have linearly separable data or continuous variable data, in the other hand, random forest classifier works better with Categorial variable data. Since we have a lot of categorial features in our dataset, in my opinion it is best to use the random forest classifier.

## Fit a LogisticRegression model and RandomForestClassifier model

```python
clf = RandomForestClassifier(n_estimators = 100) 

clf.fit(X_train, y_train)
 
# performing predictions on the test dataset
y_pred = clf.predict(X_test)
 
# using metrics module for accuracy calculation
print("ACCURACY OF THE RANDOM FOREST CLASSIFIER MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
reg = LogisticRegression().fit(X_train, y_train)

# performing predictions on the test dataset
y_pred = reg.predict(X_test)
 
# using metrics module for accuracy calculation
print("ACCURACY OF THE LOGISTIC REGRESSION MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

### results and thoughts.

On unscaled data, we can see that random forest classifier performed better and got a better score.
So our prediction turned to be true

## Revisit the Preprocessing: Scale the data

```python
# Scaling Our data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Prediction : Scaled data

After scalling the data, I think that the score of Logistic Regression will improve whereas the score for Random Forest will remain the same. Graphical-model classifiers like Random Forest are invariant to feature scaling.

```python
# Train the Logistic Regression model on the scaled data
logisticRegr_scaled = LogisticRegression().fit(X_train_scaled, y_train)
print("LogisticRegression scaled score: ", logisticRegr_scaled.score(X_test_scaled, y_test))
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
randomForestClass_scaled = RandomForestClassifier().fit(X_train_scaled, y_train)
print("RandomForestClassifier scaled score: ", randomForestClass_scaled.score(X_test_scaled, y_test))
```

### results and thoughts.

To conclude, scaling data greatly improved the score of the Logistic Regression model from .52 to .72 so that it outperformed the Random Forest classifier model. This shows that sometimes a simple model with scaled data can be a better fit than one with more complexity.
